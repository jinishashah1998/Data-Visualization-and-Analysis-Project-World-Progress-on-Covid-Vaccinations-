```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv("country_vaccinations.csv")
df
```

```python
df.drop(['source_website', 'source_name', 'daily_vaccinations_raw', 'daily_vaccinations', 'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred' ],axis='columns', inplace=True)
```

```python
df
```

```python
df.dropna(subset =['total_vaccinations'], inplace=True)
df
```

```python
us_vaccine_df = df[df['country'] == 'United States']
us_vaccine_df = us_vaccine_df.drop(['country', 'iso_code', 'vaccines'], axis='columns')
us_vaccine_df = us_vaccine_df.dropna(subset =['people_vaccinated', 'people_vaccinated_per_hundred'])
us_vaccine_df = us_vaccine_df.set_index('date')
us_vaccine_df['daily_vaccinations_per_million'] = us_vaccine_df['daily_vaccinations_per_million'].fillna(0)
us_vaccine_df
```

```python
country_df = df.groupby(['country'])[['iso_code','total_vaccinations', 'total_vaccinations_per_hundred']]
total_vaccines_df = country_df.agg('max')
total_vaccines_df
```

```python
total_vaccines_df.dtypes
```

```python
countries_df = df[['country', 'iso_code']].drop_duplicates()
countries_df = countries_df.set_index('iso_code')
countries_df
```

```python
country_vaccines_df = df[['country','iso_code','vaccines']]
country_vaccines_df = country_vaccines_df.drop_duplicates(subset=['country', 'iso_code'])
country_vaccines_df = country_vaccines_df.set_index(['country'])
country_vaccines_df
```

```python
country_vaccine_info_df = pd.merge(total_vaccines_df, country_vaccines_df, on = ["country", "iso_code"])
country_vaccine_info_df.drop('total_vaccinations_per_hundred', axis='columns', inplace=True)
country_vaccine_info_df
```

```python
people_df = df.groupby(['country'])[['iso_code','date','people_vaccinated', 'people_vaccinated_per_hundred']]
people_vaccines_df = people_df.agg('max')
people_vaccines_df = people_vaccines_df.dropna()
people_vaccines_df
```

```python
people_vaccines_df.to_csv('csv files\people_vaccines.csv')
country_vaccine_info_df.to_csv('csv files\country_vaccine_info.csv')
```

```python
total_vaccines_df.to_csv('csv files\\total_vaccines.csv')
us_vaccine_df.to_csv('csv files\\us_vaccines_info.csv')
```

```python
country_vaccines_df['vac_lst'] = country_vaccines_df['vaccines'].str.split(', ')
country_vaccines_df

new_df = pd.DataFrame()
num = 0
for i, lst in enumerate(country_vaccines_df['vac_lst']):
    for j in range(len(lst)):
        new_df[num] = [country_vaccines_df['iso_code'][i]] + [lst[j]]
        num += 1
vaccines_df = new_df.transpose()
vaccines_df = vaccines_df.rename(columns={0: "iso_code", 1: "vaccine"})
vaccines_df = vaccines_df.set_index(['iso_code', 'vaccine'])
vaccines_df
```

```python
countries_df.to_csv('csv files\\countries.csv')
vaccines_df.to_csv('csv files\\country_vaccines.csv')
```
