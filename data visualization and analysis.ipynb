```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
us_df = pd.read_csv("csv files//us_vaccines_info.csv")
us_df
```

```python
plt.figure(figsize=(20,10))
plt.plot(us_df['date'],us_df['total_vaccinations'] , color="red", label = "total vaccinations")
plt.plot(us_df['date'], us_df['people_vaccinated'],color="green", label = "people vaccinated")
plt.xticks(rotation=70)
plt.legend(loc="best")
plt.title("US Total Vaccinations vs. People Vaccinated", fontsize=20)
plt.xlabel("date", fontsize=14)
plt.ylabel("number", fontsize=14)
plt.grid()
plt.savefig("charts/us_total_vacc_chart.png")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(us_df['date'],us_df['total_vaccinations_per_hundred'] , color="red", label = "total vaccinations per hundred")
plt.plot(us_df['date'], us_df['people_vaccinated_per_hundred'],color="green", label = "people vaccinated per hundred")
plt.xticks(rotation=70)
plt.legend(loc="best")
plt.title("US Total Vaccinations Per Hundred vs. People Vaccinated Per Hundred", fontsize=20)
plt.xlabel("date", fontsize=14)
plt.ylabel("number", fontsize=14)
plt.grid()
plt.savefig("charts/us_vaccination_per_hundred.png")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(us_df['date'],us_df['daily_vaccinations_per_million'],marker = 'o',alpha = 0.8, color='black')
plt.plot(us_df['date'],us_df['daily_vaccinations_per_million'] , color="blue")
plt.xticks(rotation=70)
plt.title("Daily US Vaccinations Per Million", fontsize=20)
plt.xlabel("date", fontsize=14)
plt.ylabel("number of vaccinations", fontsize=14)
plt.grid()
plt.savefig("charts/us_daily_vacc_chart.png")
plt.show()
```

```python
x = us_df['total_vaccinations']
y = us_df['people_vaccinated']
m, b = np.polyfit(x, y, 1)
m = round(m,2)
b=round(b,2)

plt.figure(figsize=(15,10))
plt.scatter(x,y,marker = 'o',alpha = 0.8, color='red')
plt.plot(x, m*x + b, c='black', label=str(m)+"*x +"+str(b))
plt.text(85000000, 70000000, str(m)+"*x +"+str(b), fontsize=12)
plt.title("US Vaccinations (Total Vaccinations Vs People Vaccinated)", fontsize=18)
plt.xlabel("total vaccinations", fontsize=12)
plt.ylabel("number of people vaccinated", fontsize=12)
plt.grid()
plt.savefig("charts/lin_reg_us_vacc.png")
plt.show()
```

```python
total_vacc_df = pd.read_csv("csv files//total_vaccines.csv")
sorted_df = total_vacc_df.sort_values('total_vaccinations', ascending= False)
top_ten_df = sorted_df.head(10)
top_ten_df
```

```python
colours = ['red', 'darkorange', 'gold', 'green', 'deepskyblue', 'blue', 'indigo', 'purple', 'violet', 'deeppink']

plt.figure(figsize=(10,5))
plt.barh(top_ten_df['country'], top_ten_df['total_vaccinations'], color=colours)
plt.title('Total Vaccinations for Top 10 Countries', fontsize=16)
plt.ylabel('country', fontsize=12)
plt.xlabel('vaccinations', fontsize=12)
plt.savefig("charts/top_ten_vacc.png")
plt.show()
```

```python
vaccine_count_df = pd.read_csv("csv files/final_vaccine_count.csv")
vaccine_count_df
```

```python
plt.figure(figsize=(20,5))
plt.bar(vaccine_count_df['VACCINE_NAME'],vaccine_count_df['COUNTRY_COUNT'], color='purple')
for index, value in enumerate(vaccine_count_df['COUNTRY_COUNT']):
    plt.text(index, value, str(value), ha='center', va='baseline', fontsize=12)
plt.title("Types of Vaccines and their Country Counts", fontsize=16)
plt.ylabel('number of countries', fontsize=12)
plt.xlabel('vaccine type', fontsize=12)
plt.savefig("charts/vacc_country_counts.png")
plt.show()
```
